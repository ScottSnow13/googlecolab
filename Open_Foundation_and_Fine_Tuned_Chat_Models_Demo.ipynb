```python
import lamini
from openai import OpenAI

from google.colab import userdata
gpt_key = userdata.get('OpenAI_API_KEY')

from google.colab import userdata
llama_key = userdata.get('LAMINI_API_KEY')

lamini.api_key = llama_key

client = OpenAI(
    api_key = gpt_key
)
```

chatgpt.png

```python
completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "user", "content": "What’s a good haircut that looks great on everybody?"}
  ]
)

response = completion.choices[0].message.content

print(response)
```
